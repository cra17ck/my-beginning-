```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
dictionary = {'Модель':[],'Комплектация':[], 'Город':[], 'Объём_двигателя':[], 'Пробег':[],'Топливо':[],'Трансмиссия':[],'Привод':[],'Цена':[],'Город':[], 'Дата':[]}

```

```python
for i in range(1,8):
    print(f'https://auto.drom.ru/toyota/mark_ii/page{i}/?transmission[]=1&unsold=1')
```

```python
bull_count = 1
for i in range(1,8):
    url = f'https://auto.drom.ru/toyota/mark_ii/page{i}/?transmission[]=1&unsold=1'
    # url = f'https://auto.drom.ru/toyota/mark_ii/page1/?transmission[]=1&unsold=1'
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')

    news_items = soup.find_all(class_='css-4zflqt e1huvdhj1')

    for css in news_items:

        model = css.find(class_='css-16kqa8y e3f4v4l2').text
        equipment = css.find(class_='css-1hd50jd e3f4v4l0')
        if equipment is not None:
            equipment = equipment.text
        else:
            equipment = "Не указано"
        price = css.find(class_='css-46itwz e162wx9x0').text
        hihi = css.find_all(class_='css-1x4jcds eotelyr0')
        for haha in hihi:
            city = haha.find(attrs={"data-ftid": "bull_location"}).text
            date = haha.find(attrs={'data-ftid': 'bull_date'}).text
            
        
        car_info = css.find_all(class_='css-1fe6w6s e162wx9x0')
        for i in car_info:
            car_det = i.text.split(',')
            dvigatel = car_det[0]
            fuel = car_det[1]
            trans = car_det[2]
            drive = car_det[3]  
            try:
                probeg = car_det[4]
            except IndexError:
                probeg = 'Не указано'


        dictionary['Модель'].append(model)
        dictionary['Комплектация'].append(equipment)
        dictionary['Объём_двигателя'].append(dvigatel)
        dictionary['Пробег'].append(probeg)
        dictionary['Топливо'].append(fuel)
        dictionary['Трансмиссия'].append(trans)
        dictionary['Привод'].append(drive)
        dictionary['Цена'].append(price)
        dictionary['Город'].append(city)
        dictionary['Дата'].append(date)
        print(f"Обьявление №{bull_count}")
        print(f'Модель: {model}\nКомплектация: {equipment}\nГород: {city}\nОбъем двигателя: {dvigatel}\nПробег: {probeg}\nЦена: {price}\nТопливо: {fuel}\nТрансмиссия: {trans}\nПривод: {drive}\nДата: {date}\n\n\n')
        bull_count += 1

```

```python
mark_df = pd.DataFrame.from_dict(dictionary)
```
